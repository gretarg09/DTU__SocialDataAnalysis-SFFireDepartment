### Reading in the fire station location data and creating a json file for the javascript code

```python
# lets start by getting the data
import csv
import math
import matplotlib.pyplot as plt
from collections import defaultdict
from collections import Counter

data = []

def process(station,location,latitude,longitude,battalion,division):
    
    if(latitude is not ""):
        data.append({
                "station":station,
                "location":location,
                "lat":float(latitude),
                "lon":float(longitude),
                "battalion":battalion,
                "division":division
            })
    
with open("csv_files/firestations_location.csv", "rb") as csvfile:
    reader = csv.DictReader(csvfile,delimiter=",")
    for row in reader:
        station = row["station"] 
        location = row["location"]
        latitude = row["latitude"]
        longitude = row["longitude"]
        battalion = row["battalion"]
        division = row["division"]
        
        process(station,location,latitude,longitude,battalion,division)
```

```python
# Creating json file for visualization 2 to import into our javascript code
import json
with open('Data/fire_station.json', 'w') as fp:
    json.dump(data, fp)
```
