### Reading in the neighborhood statistic data and creating a json file for the javascript code

```python
# lets start by getting the data
import csv
import math
import matplotlib.pyplot as plt
from collections import defaultdict
from collections import Counter

# Dataset url:
# https://data.sfgov.org/Public-Safety/Fire-Department-Calls-for-Service/nuek-vuh3

data = []

def process(neighborhood,total_population,people_density,area,household_income, poverty_level, foreign_born_residents):    
    data.append({
            "neighborhood":neighborhood,
            "total_population":total_population,
            "people_density":people_density,
            "area":area,
            "household_income":household_income,
            "poverty_level": poverty_level,
            "foreign_born_residents" : foreign_born_residents
        })
    
with open("csv_files/neighborhood_statistic.csv", "rb") as csvfile:
    reader = csv.DictReader(csvfile,delimiter=";")
    for row in reader:
        neighborhood = row["Neighborhood"] 
        total_population = row["Total Population"]
        people_density = row["People Per Square Mile"]
        area = row["Area Square Mile"]
        household_income = row["Median Household Income "]
        poverty_level = row["ratio of people below poverty level"]
        foreign_born_residents = row["ratio of foreign born residents"]
        
        process(neighborhood,total_population,people_density,area,household_income, poverty_level, foreign_born_residents)
        
neighborhood_filter = ["Castro","Upper Market","Financial District","South Beach",
                       "Oceanview","Merced Heights","Merced Manor","Ingleside", 
                       "Sunset ","Parkside"]

data = [line for line in data if line["neighborhood"] not in neighborhood_filter]
```

```python
# Creating json file for visualization 2 to import into our javascript code
import json
with open('neighborhood_statistic.json', 'w') as fp:
    json.dump(data, fp)
```
